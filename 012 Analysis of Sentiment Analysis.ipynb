```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
plt.style.use('grayscale')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['figure.dpi'] = 1000
```

```python
df = pd.read_csv('sentiment data/crowdflower-sentiment-analysis-in-text/data/text_emotion.csv')
```

```python
list(df.sentiment.unique())
```

```python
plt.rcParams['figure.figsize'] = (16,2.5)
plt.rcParams['font.size'] = 16

for sentiment in df.sentiment.unique():
    plt.bar(sentiment, len(df[df.sentiment == sentiment]), color = 'C0')

plt.title('Distribution of the Label on the Crowdflower\'s data set', fontweight='bold')
plt.xticks(rotation=30)
plt.savefig('img/012-01DistributionInLabel.eps', format = 'eps', bbox_inches='tight')
```

```python
tdf = pd.read_csv('misc/results_evaluation.csv')
```

```python
tdf = tdf.sort_values('fee_cleaned').tail(50).sort_values('player_name')
```

```python
df = pd.read_csv('sentiment data/sentiment_results.csv')
```

```python
df['Transfer Fee'] = tdf.fee_cleaned.values
df['StackingNN'] = tdf.NNstack.values
```

```python
df
```

```python
sentiments = ['empty', 'sadness', 'enthusiasm', 'neutral', 'worry', 'surprise', 'love', 'fun', 'hate', 'happiness', 'boredom', 'relief', 'anger']
```

```python
for sentiment in sentiments:
    print(sentiment, 'min', df[df[sentiment] == df[sentiment].min()]['player'].iloc[0])
    print(sentiment, 'max', df[df[sentiment] == df[sentiment].max()]['player'].iloc[0])
```

```python
df.corr()
```

```python
plt.style.use('grayscale')
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['figure.dpi'] = 1080
plt.rcParams['figure.figsize'] = (4,6)

sns.heatmap(df.corr()[['Transfer Fee', 'StackingNN']].drop(['Transfer Fee', 'StackingNN', 'Unnamed: 0']), vmin=-.5, vmax=.5, cmap='binary', linewidths=1, linecolor='black', annot = True, cbar = False)

plt.title('Correlations of Sentiments with Transfer Fees', fontweight='bold')
plt.savefig('img/012-2SentCorrs.eps', format='eps', bbox_inches='tight')
```

```python
best_n_worst_performers = [["empty","Jorginho to Chelsea FC","Cristiano Ronaldo to Juventus FC"],
["sadness","Alexandre Lacazette to Arsenal FC","Ángel Di María to Manchester United"],
["enthusiasm","Jorginho to Chelsea FC","Mateo Kovacic to Chelsea FC"],
["neutral","Hulk to shanghai SIPG","Ferland Mendy to Real Madrid"],
["worry","Fabinho to Liverpool FC","João Félix to Atlético Madrid"],
["surprise","Neymar to Paris Saint-Germain","Raheem Sterling to Manchester City"],
["love","Ferland Mendy to Real Madrid","Kyle Walker to Manchester City"],
["fun","Neymar to Paris Saint-Germain","Hulk to Shanghai SIPG"],
["hate","Romelu Lukaku to Manchester United","Raheem Sterling to Manchester City"],
["happiness","Harry Maguire to Manchester United","Bernardo Silva to Manchester City"],
["boredom","João Félix to Atlético Madrid","Alexandre Lacazette to Arsenal FC"],
["relief","Álvaro Morata to Atlético Madrid","Mateo Kovacic to Chelsea FC"],
["anger","Nathan Aké to Manchester City","Fabinho to Liverpool FC"]]
bnw_df = pd.DataFrame(best_n_worst_performers, columns=['sentiment','Worst Performer','Best Performer'])
bnw_df = bnw_df.set_index('sentiment')
```

```python
bnw_df = df.corr().join(bnw_df)
```
